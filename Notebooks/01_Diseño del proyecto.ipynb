El **primer paso** es cuando un usuario llega a la web del ecommerce. Normalmente vendrá desde:

* Campañas de pago: Paid ads como Facebook Ads o Google Ads.
* Contenido orgánico: Blog, rrss, ...
* Tráfico directo: Conoce la url y la introduce en el navegador.

Ese tráfico se llama visitas, y las páginas que van viendo se llaman páginas vistas. En este proyecto las denominaremos como views.

En el **segundo paso**, el usuario navega por la web y cuando le gusta un producto lo mete en el carrito.

* Puede sacar productos del carrito, salir sin comprar nada, o finalmente hacer el pedido.
* Un proceso común es la venta cruzada, en la cual se recomiendan al usuario otros productos que también podrían interesarle.

El **tercer paso** sería la recaptación del cliente mediante retargeting o email marketing.

Todo este proceso se denomina **funnel** o **customer journey**. Otras métricas de interés para gestionar este tipo de negocios son:

* *CPA (Cost Per Adquisition):* Coste agregado de captación de un cliente a través de una campaña de marketing.
* *AOV (Average Order Value):* Valor medio del carrito de compra de los clientes, nos informa acerca de la cantidad de dinero que gastan los usuarios como promedio en el ecommerce.
* *Frecuencia de compra:* Número de veces que un usuario compra en el ecommerce.
* *LTV (Lifetime Value):* Permite predecir el beneficio medio de un cliente a lo largo de toda su vida como comprador en el ecommerce.
* *Churn rate (Tasa de cancelación):* Métrica que informa al ecommerce sobre el número de clientes que ya no compran sus servicios.

Para incrementar los beneficios de un negocio, podemos trabajar sobre tres conceptos clave:

* **Más clientes:** Esto implica conseguir más visitas y mayor conversión.
* **Más frecuencia:** Esto implica conseguir que los mismos clientes compren más veces.
* **Mayor ticket medio:** Esto implica conseguir que se compre más o más caro en la misma sesión de compra.

Por tanto, las palancas operativas sobre las que trabajar serán las siguientes:

* **Customer journey:** Cómo podemos optimizar cada uno de los pasos del proceso.
* **Clientes:** Cómo podemos usar la información disponible de los clientes para optimizar las campañas que realicemos.
* **Productos:** Cómo podemos optimizar el catálogo de productos e identificar de manera personalizada qué productos tenemos que poner delante de cada cliente.
